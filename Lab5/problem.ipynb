```python
from sklearn.mixture import GaussianMixture
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
img_path = "./zebra.jpg"
org_img = cv2.imread(img_path)
org_img = cv2.cvtColor(org_img, cv2.COLOR_BGR2RGB)
plt.imshow(org_img)
plt.axis("off")
plt.title(f"Zebra: {org_img.shape}")
plt.show()
```

```python
h, w = org_img.shape[:2]
img = org_img.reshape((-1, 3))
img.shape
```

```python
gm = GaussianMixture(n_components=2, covariance_type="full", random_state=23520442)
gm.fit(img)
```

```python
obj = gm.predict(img)
obj = obj.reshape(h, w, -1)
plt.imshow(obj)
plt.axis("off")
plt.title("Mask")
plt.show()
```

```python
segmented = np.where(obj!=0, 255, org_img).astype(np.uint8)
plt.imshow(segmented)
plt.axis("off")
plt.title("Segmented object")
plt.show()
```
